```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Bollywood.csv")

# Display first 5 rows
df.head()
```

```python
print(df.isnull().sum())
```

```python
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df
```

```python
df.info()
df
```

**Statistical Summary**

```python
df.describe()
df
```

**Budget vs. Box Office Collection (Insight #1)**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Budget'], df['BoxOfficeCollection'], color='blue', alpha=1)
plt.xlabel("Budget (in Crores)")
plt.ylabel("Box Office Collection (in Crores)")
plt.title("Budget vs. Box Office Collection")
plt.grid(True)
plt.show()
```

**Effect of Release Time on Box Office Collection (Insight #2)**

```python
plt.figure(figsize=(8,5))
release_times = df.groupby('ReleaseTime')['BoxOfficeCollection'].mean()
plt.bar(release_times.index, release_times.values, color=['red', 'green', 'blue', 'orange'])
plt.xlabel("Release Time")
plt.ylabel("Average Box Office Collection")
plt.title("Effect of Release Time on Box Office Collection")
plt.grid(axis='y')
plt.show()
```

**Popular Movie Genres (Insight #3)**

```python
plt.figure(figsize=(8,5))
genre_counts = df['Genre'].value_counts()
colors = ["Green", "orange", "grey", "pink", "red"]
genre_colors = colors[:len(genre_counts)]
plt.bar(genre_counts.index, genre_counts.values, color=genre_colors)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Movie Genre Distribution")
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

**YouTube Views vs. Box Office Collection (Insight #4)**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['YoutubeViews'], df['BoxOfficeCollection'], color='blue', alpha=0.5, label="Movies")
m, b = np.polyfit(df['YoutubeViews'], df['BoxOfficeCollection'], 1)  # Linear regression
plt.plot(df['YoutubeViews'], m*df['YoutubeViews'] + b, color='red', linewidth=2, label="Trend Line")
plt.xlabel("YouTube Views")
plt.ylabel("Box Office Collection")
plt.title("YouTube Views vs. Box Office Collection")
plt.legend()
plt.grid(True)
plt.show()
```

**YouTube Like-Dislike Ratio vs. Collection (Insight #5)**

```python
plt.figure(figsize=(8, 5))
df['Like_Dislike_Category'] = pd.qcut(df['Like_Dislike_Ratio'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
avg_box_office = df.groupby('Like_Dislike_Category', observed=True)['BoxOfficeCollection'].mean().sort_index()
colors = ['red', 'orange', 'blue', 'green']
bars = plt.bar(avg_box_office.index, avg_box_office.values, color=colors)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
             f"{bar.get_height():,.0f}", ha='center', va='bottom', fontsize=11, fontweight='bold')
plt.xlabel("Like-Dislike Ratio Category")
plt.ylabel("Average Box Office Collection")
plt.title("Average Box Office Collection by Like-Dislike Ratio")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Relationship Between Budget and YouTube Views (Insight#6)**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Budget'], df['YoutubeViews'], color='brown', alpha=0.6)
plt.xlabel("Budget (in Crores)")
plt.ylabel("YouTube Views")
plt.title("Budget vs. YouTube Views")
plt.grid(True)
plt.show()
```

**Distribution of Box Office Collection (Insight#7)**

```python
plt.figure(figsize=(8,5))
plt.hist(df['BoxOfficeCollection'], bins=15, color='Black', edgecolor='black', alpha=0.7)
plt.xlabel("Box Office Collection (in Crores)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Box Office Collection")
plt.grid(axis='y')
plt.show()
```

**Monthly Trend of Movie Releases (Insight#8)**

```python
monthly_releases = df['ReleaseMonth'].value_counts().sort_index()
peak_month = monthly_releases.idxmax()
peak_month_count = monthly_releases.max()
colors = ['blue'] * 12 
colors[peak_month - 1] = 'red'  
plt.figure(figsize=(10,6))
bars = plt.bar(monthly_releases.index, monthly_releases.values, color=colors, edgecolor='black', alpha=0.8)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Movies Released", fontsize=12)
plt.title("Monthly Trend of Bollywood Movie Releases (2013-2015)", fontsize=14, fontweight='bold')
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], fontsize=10)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.3, str(height), ha='center', fontsize=10, fontweight='bold')
plt.annotate(f"Peak Release Month: Jan ({peak_month_count} movies)",
             xy=(peak_month, peak_month_count), xytext=(peak_month+2, peak_month_count - 2),
             arrowprops=dict(facecolor='red', shrink=0.05), fontsize=12, fontweight='bold', color='red')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Summary of Insights:
### Insight #1: Budget vs. Box Office Collection

- The most common genre in this dataset is **Comedy**, followed by **Action**.

### Insight #4: YouTube Views vs. Box Office Collection
- Higher YouTube trailer views generally correlate with better box office success.

### Insight #5: Like-Dislike Ratio vs. Box Office Collection
- Movies with a **higher like-dislike ratio** tend to perform better at the box office.

### Insight #6: Budget vs. YouTube Views
- Big-budget movies tend to get more views on YouTube, but some low-budget films also gain high traction.

### Insight #7: Distribution of Box Office Collection
- The **majority** of Bollywood movies earn **less than 50 Crores**, while a few blockbusters cross 200 Crores.

### Insight #8: Monthly Trend of Movie Releases
- The **January** witnessed the highest number of Bollywood movie releases between **2013-2015**, indicating a strategic preference for **New Year and winter holiday releases.**
